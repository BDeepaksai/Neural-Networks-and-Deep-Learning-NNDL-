```python
import ssl, certifi, os
os.environ['SSL_CERT_FILE'] = certifi.where()
ssl._create_default_https_context = ssl._create_unverified_context

import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# ---------- CONFIG ----------
DATA_DIR = "images"                                # <-- FIXED
TRAIN_DIR = os.path.join(DATA_DIR, "train")        # <-- FIXED
TEST_DIR = os.path.join(DATA_DIR, "validation")    # <-- FIXED
MODEL_OUT = "models/emotion_model.h5"
IMG_SIZE = 96
BATCH_SIZE = 32
EPOCHS = 20

EMOTION_LABELS = []


# ---------- DATA PIPELINE ----------
def load_datasets():

    datagen_train = ImageDataGenerator(
        rescale=1./255,
        rotation_range=20,
        zoom_range=0.2,
        horizontal_flip=True
    )
    datagen_test = ImageDataGenerator(rescale=1./255)

    train_data = datagen_train.flow_from_directory(
        TRAIN_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        color_mode="rgb",
        class_mode="categorical",
        shuffle=True
    )

    test_data = datagen_test.flow_from_directory(
        TEST_DIR,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        color_mode="rgb",
        class_mode="categorical",
        shuffle=False
    )

    # Auto-detect classes
    global EMOTION_LABELS
    EMOTION_LABELS = list(train_data.class_indices.keys())

    print("\n Detected Classes:", EMOTION_LABELS)
    print(" Number of classes =", len(EMOTION_LABELS))

    return train_data, test_data


# ---------- MODEL ----------
def build_model():

    num_classes = len(EMOTION_LABELS)

    base_model = tf.keras.applications.MobileNetV2(
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        include_top=False,
        weights="imagenet"
    )
    base_model.trainable = False

    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation="softmax")
    ])

    model.compile(
        optimizer=optimizers.Adam(1e-3),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )

    print("\n Model Built Successfully (Classes:", num_classes, ")")

    return model


# ---------- TRAIN ----------
def train_model():

    os.makedirs("models", exist_ok=True)
    os.makedirs("results", exist_ok=True)

    train_data, test_data = load_datasets()
    model = build_model()

    ckpt_cb = callbacks.ModelCheckpoint(
        MODEL_OUT, monitor="accuracy", save_best_only=True, verbose=1
    )

    # Train
    history = model.fit(
        train_data,
        epochs=EPOCHS,
        callbacks=[ckpt_cb],
        verbose=1
    )

    # Plot Accuracy & Loss
    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="train_acc")
    plt.title("Training Accuracy")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="train_loss")
    plt.title("Training Loss")
    plt.legend()

    plt.tight_layout()
    plt.savefig("results/accuracy_loss_plot.png")
    print("\n Saved training graph to results/accuracy_loss_plot.png")

    # Test Evaluation
    test_loss, test_acc = model.evaluate(test_data)
    print(f" Test Accuracy: {test_acc*100:.2f}%")

    model.save(MODEL_OUT)
    print(f"\nðŸ’¾ Model saved to {MODEL_OUT}")


if __name__ == "__main__":
    train_model()
```
